### plotting the correlation between our data features to see if the remaining features will be a good choice for our "Linear model"

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12}

        )
    
plot_correlation_map(data)
```

### we see that the "price" feature is highly correlated with the "sqft_living" ,"sqft_above" and" bathrooms" features <br>so they are highly linearly related with the price and that strengthen our 'linear" model<br> a natural approches will be to use only these 3 feature in our  "Linear " model to get the best performance

### spliting our data into train/test 

```python
x = data.iloc[:,1:]
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.25 , random_state = 30 )
```

### here we are creating  2 empty narrays that will include all the "MSE" and "R squared" of both "Linear" and "Multi-Linear" in order to compare them

```python
MSE= np.array(())
R2 = np.array(())
```

### applying "linear regression" to each feature

```python
figx ,axx = plt.subplots(nrows = 3 ,ncols = 1 ,figsize=(15,14))

# 3 since they are 3 features
for i in range(3):
    # the "z" variable will select our feature while iterating
    if i==0:
        z = "bathrooms"
    elif i == 1:
        z = "sqft_living"
    elif i ==2:
        z ="sqft_above"
    
    # will used for reshaping our 1-D narray
    len_0 = len(x_train[z])
    len_1 = len(x_test[z])
    
    # iterating over  x_train/x_test of our features
      # the "to_numpy() method is used to transform our pandas series to an narray"
      # the "reshape"  method is used to transform our 1D array to a 2d one to be fitted on the model     
    X_train = x_train[z].to_numpy().reshape(len_0,1)
    X_test = x_test[z].to_numpy().reshape(len_1,1)
    
    # creating our model + fitting the training data
    model0 = LinearRegression()
    model0.fit(X_train,y_train)
    
    # the prediction of our model
    pred = model0.predict(X_test)
    
    #plotting
    axx[i].plot(X_test,pred , color = 'k')
    axx[i].scatter(X_test,y_test)
    axx[i].set_xlabel(z,fontsize = 18)
    axx[i].set_ylabel("price", fontsize = 18)
    
    #calculating the "MSE" and "R squared" of the model
    mse = mean_squared_error(y_test,pred)
    r2 = metrics.r2_score(y_test,pred)
    
    #plotting
    axx[i].text( x = 0.01 , y = 0.78,s = "MSE = {}\nR2 = {}".format(mse,r2),
                bbox = dict(boxstyle = 'round',facecolor = '#bff4ff' , alpha = 0.4),
                transform=axx[i].transAxes,
                    fontsize=20 )
    axx[i].text( x = 0.01 , y = 0.5,s = "a = {}\nb = {}".format(model0.coef_[0],model0.intercept_),
                bbox = dict(boxstyle = 'round' ,facecolor = '#bff4ff' , alpha = 0.4),
                transform=axx[i].transAxes,
                    fontsize=20 )
    
    #appending the "MSE" and "R squared" to  the MSE and R2 narray defined above
    MSE = np.append(MSE,mse)
    R2 = np.append(R2,r2)
```

### applying "Polynomial Regression" in each feauture

```python
#creating a new figures with its axes
 #here we are using two figures just for a clearer outcome
figu ,axu = plt.subplots(nrows = 4 , ncols =1  , figsize=(15,18))
figuu,axuu = plt.subplots(nrows = 4 , ncols =1  , figsize=(15,18))

for i in range(8):
    #this variable will be use throw the code just for iterating over the two figures and their axes
    i_ = i
    j = i//4
    i = i%4 
    
    # toggling from the first figure to the second
    if j==0:
        axes = axu
    else:
        axes=axuu
       
    
    #color of the scatter plot and the text_box
    colord = "#3d8bff" if i_<=3 else "#ff9900"
    text_color = "#bff4ff" if i_<=3 else "#fcb74e"
    
    #creating the model
    model2 = LinearRegression()
    
    #setting the degree of our "polynomial regression"
    poly = PolynomialFeatures (degree = degree[i_])
    
    
    #setting our x_train/x_test + fitting it to the model
    X_train = x_train.iloc[:,[i_]].to_numpy()
    X_train_ = poly.fit_transform(X_train)
    
    X_test = x_test.iloc[:,[i_]].to_numpy()
    X_test_ = poly.fit_transform(X_test)
    
    model2.fit(X_train_,y_train)
    
    
    #prediction
    pred = model2.predict(X_test_)
    
    
    #plotting
    axes[i].plot(X_test,pred , color = 'k')
    axes[i].scatter(X_test,y_test , color = colord)
    
        
    #calculating the "MSE" and "R squared" of the model    
    mse = mean_squared_error(y_test,pred)
    r2 = metrics.r2_score(y_test,pred)
    
    #adding some text
    axes[i].text( x = 0.01 , y = 0.81 ,
              s = "MSE  = {}\nR2 = {}".format(mse,r2),
              transform = axes[i].transAxes,
              bbox = dict(boxstyle = 'round',facecolor = text_color , alpha = 0.5),
              fontsize = 18)
    
    
    #appending the "MSE" and "R squared" to  the MSE and R2 narray defined above
    poly_MSE = np.append(poly_MSE ,mse)
    poly_R2 = np.append(poly_R2 ,r2)
```
